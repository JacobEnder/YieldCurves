### Load, clean and construct initial Duration and YtM Data

```python
from scipy.interpolate import CubicSpline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

row = [0,4]
durations = [0.25, 0.5, 1, 2, 3, 5, 7, 10, 30]

def p2f(x):
    return float(x.strip('%'))/100

three_m = pd.read_csv('3M.csv', usecols= ['Date', 'Close'], skiprows = lambda x : x not in row, converters={'Close':p2f})
six_m = pd.read_csv('6M.csv', usecols= ['Date', 'Close'], skiprows = lambda x : x not in row, converters={'Close':p2f})
one_y = pd.read_csv('1Y.csv', usecols= ['Date', 'Close'], skiprows = lambda x : x not in row, converters={'Close':p2f})
two_y = pd.read_csv('2Y.csv', usecols= ['Date', 'Close'], skiprows = lambda x : x not in row, converters={'Close':p2f})
three_y = pd.read_csv('3Y.csv', usecols= ['Date', 'Close'], skiprows = lambda x : x not in row, converters={'Close':p2f})
five_y = pd.read_csv('5Y.csv', usecols= ['Date', 'Close'], skiprows = lambda x : x not in row, converters={'Close':p2f})
seven_y = pd.read_csv('7Y.csv', usecols= ['Date', 'Close'], skiprows = lambda x : x not in row, converters={'Close':p2f})
ten_y = pd.read_csv('10Y.csv', usecols= ['Date', 'Close'], skiprows = lambda x : x not in row,converters={'Close':p2f})
thirty_y = pd.read_csv('30Y.csv', usecols= ['Date', 'Close'], skiprows = lambda x : x not in row, converters={'Close':p2f})

x1 = np.array(three_m.Close)
x2 = np.append(x1, six_m.Close)
x3 = np.append(x2, one_y.Close)
x4 = np.append(x3, two_y.Close)
x5 = np.append(x4, three_y.Close)
x6 = np.append(x5, five_y.Close)
x7 = np.append(x6, seven_y.Close)
x8 = np.append(x7, ten_y.Close)

yields = np.append(x8, thirty_y.Close)

d = {'Duration': durations, 'Yield': yields}

data = pd.DataFrame(data=d)
data




```

```python
f = CubicSpline(durations, yields, bc_type="natural")
xvals = np.linspace(0,30,121)
interpolate_y = f(xvals)

interpolated_raw = {'Duration': xvals, 'Yield': interpolate_y}

inter_data = pd.DataFrame(data=interpolated_raw)
inter_data
```

```python
cds = 0.0049
```

```python
anchor_points = [1,2,3,4,8,12,20,28,40,60,80,100]

durations_to_interpolate = []
yields_to_interpolate = []

for i in anchor_points:
    durations_to_interpolate.append(inter_data.Duration[i])
    yields_to_interpolate.append(inter_data.Yield[i]+cds)

durs = np.array(durations_to_interpolate)
yields = np.array(yields_to_interpolate)

e = {'Duration': durs, 'Yield': yields}

df = pd.DataFrame(data=e)
to_spline = df.copy()
to_spline['RF'] = round(df['Yield']*100, 5)
df
```

```python
durations_to_interpolate = np.array(to_spline.Duration)
yields_to_interpolate = np.array(to_spline.Yield)
rf = np.array(to_spline.RF)

f = CubicSpline(durations_to_interpolate, yields_to_interpolate, bc_type="natural")

x_space = np.linspace(0.25, 25, 100)
y_splined = f(x_space)

new_space = {'Period': x_space, 'Interest': y_splined}
df1 = pd.DataFrame(data=new_space)
df1['RF'] = round(df1['Interest']*100,5)
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

M0 = np.min(df1.RF) - 0.07
M1 = np.max(df1.RF) + 0.07

fontsize=13

fig, ax = plt.subplots(figsize=(13,7))

dur = [0.25, 0.5, 1., 2., 3., 5., 7., 10., 30.]
df1["Observed"] = df1["RF"].isin(dur)

sns.scatterplot(df = df1, x = "Period", y = "RF", hue = "Observed", ax = ax)
sns.lineplot(df = df1, x = "Period", y = "RF", ax = ax)

plt.xlabel('Period',fontsize=fontsize)
plt.ylabel('Yield',fontsize=fontsize)

plt.xlim(0, 25)
plt.ylim(M0, M1)
plt.legend(loc="center right")
plt.title("USD Treasury Bill Yield Curve - Cubic Spline Interpolation",fontsize=fontsize)

plt.show()
```
