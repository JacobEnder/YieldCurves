# Functions

```python
from scipy.interpolate import CubicSpline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dir = "./data/"
files = os.listdir(dir)
durations = [0.25, 0.33, 0.5, 1, 2, 3, 5, 7, 10, 30]
```

```python
def p2f(x):
    return float(x.strip('%'))/100

def construct_yield_df(i=4):
    
    row = [0,i]
    yields = []
    for f in files:
        dta = pd.read_csv(dir + f, usecols= ['Date', 'Close'], skiprows = lambda x : x not in row, converters={'Close':p2f})
        yields.append(dta["Close"])

    yields = np.array(yields).flatten()
    d = {'Duration': durations, 'Yield': yields}

    return pd.DataFrame(d)

def get_interpolation_points(yield_curve):

    f = CubicSpline(yield_curve["Duration"], yield_curve["Yield"], bc_type="natural")
    n_dur = len(durations)
    n_inter_per_year = 2
    xvals = np.array([])

    for i in range(n_dur - 1):
        time_elapsed = (durations[i+1] - durations[i])
        points = np.ceil(time_elapsed * n_inter_per_year)
        ls = np.linspace(durations[i], durations[i+1], int(points), endpoint = False)
        xvals = np.concatenate((xvals, ls))
    xvals[-1] = durations[-1]

    interpolate_y = f(xvals)
    interpolated_raw = {'Duration': xvals, 'Yield': interpolate_y}

    return pd.DataFrame(interpolated_raw)

def plot_curve(yield_curve, observed_dur):

    yield_curve["RF"] = yield_curve["Yield"] * 100
    M0 = np.min(yield_curve.RF) - 0.07
    M1 = np.max(yield_curve.RF) + 0.07

    fontsize=13

    fig, ax = plt.subplots(figsize=(13,7))

    dur = [0.25, 0.5, 1., 2., 3., 5., 7., 10., 30.]
    yield_curve["Observed"] = yield_curve["Duration"].isin(observed_dur)
    yield_curve["Observed"] = yield_curve["Observed"].apply(lambda x: "Observed" if x else "Interpolated")

    sns.scatterplot(data = yield_curve, x = "Duration", y = "RF", hue = "Observed", ax = ax)
    sns.lineplot(data = yield_curve, x = "Duration", y = "RF", ax = ax)

    plt.xlabel('Duration',fontsize=fontsize)
    plt.ylabel('Yield',fontsize=fontsize)

    plt.xlim(0, 25)
    plt.ylim(M0, M1)
    plt.legend(loc="center right")
    plt.title("USD Treasury Bill Yield Curve - Cubic Spline Interpolation",fontsize=fontsize)

    plt.show()
```

# Plot One Curve

```python
curve = construct_yield_df()
curve_inter = get_interpolation_points(curve)
```

```python
plot_curve(curve_inter, curve["Duration"])
```

# Compare in Time

```python
end_date = 20
dates = range(1, end_date+1)
true_yield = np.zeros(end_date)
estimated_yield = np.zeros(end_date)
drop_idx = 4
drop_dur = durations[drop_idx]

for d in dates:
    curve = construct_yield_df(d)
    true_yield[d - 1] = curve["Yield"].iloc[drop_idx]
    curve = curve.drop(drop_idx)
    curve_inter = get_interpolation_points(curve)
    estimated_yield[d - 1] = curve_inter[curve_inter["Duration"] == drop_dur]["Yield"]
```

```python
dates = pd.read_csv(dir + "1_1M.csv", usecols= ['Date'])["Date"]
df = pd.DataFrame({"true": true_yield, "estimate": estimated_yield}, index = dates[:end_date]) * 100
df.index = pd.to_datetime(df.index)
```

```python
with plt.style.context("seaborn-v0_8-pastel", after_reset = True):
    f, ax = plt.subplots(figsize=(15, 5))
    sns.lineplot(data = df, x = df.index, y = "true", label = "True Yield", ax = ax)
    sns.lineplot(data = df, x = df.index, y = "estimate", label = "Estimated Yield", ax = ax)
    plt.xlabel("Date")
    plt.ylabel("Yield (%)")
    plt.title("2Y US Treasury Yield")
    plt.show()
```
